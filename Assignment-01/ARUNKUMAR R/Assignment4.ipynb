```python

from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()


label = le.fit_transform(df['v1'])

label
```

```python

df.drop("v1", axis=1, inplace=True)


df["v1"] = label

# printing Dataframe
df
```

```python
train_set = df.iloc[:,4:5].values
train_set
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()
scaled_tr = sc.fit_transform(train_set)
scaled_tr
```

```python
scaled_tr[60-60:60,0]
```

```python
x_train = []
y_train = []

for i in range(60,1258):
  x_train.append(scaled_tr[i-60:i,0])
  y_train.append(scaled_tr[i,0])
```

```python
x_train
```

```python
y_train
```

```python
x_train,y_train = np.array(x_train),np.array(y_train)
```

```python
x_train
```

```python
x_train.shape
```

```python
x_train = np.reshape(x_train,(1198, 60, 1))
x_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
model = Sequential()
model.add(LSTM(50, input_shape = (60,1),return_sequences=True))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50,return_sequences=True))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x_train,y_train,epochs=24,batch_size=32)
```

```python
model.save('ARK.h5')
```
